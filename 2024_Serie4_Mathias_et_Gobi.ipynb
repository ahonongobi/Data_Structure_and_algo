## Exercice 1
Concevez un algorithme pour **√©num√©rer tous les arbres binaires** d'une taille N donn√©e, o√π N correspond au nombre de noeuds (vertex) ; le N ne prend pas en compte les feuilles. Les noeuds feuilles sont repr√©sent√©s par ceci "(..)": une parenth√®se ouverte '(', deux points '..' - repr√©sentant les deux feuilles- et une parenth√®se ferm√©e ')'.

Le programme final doit afficher la liste des arbres binaires de la taille sp√©cifi√©e, un arbre par ligne. Pour N=3, il y a un total de 5 arbres binaires. L'output doit √™tre une s√©quence de parenth√®ses et points, comme ceci:

    (((..).).)
    ((.(..)).)
    ((..)(..))
    (.((..).))
    (.(.(..)))
    
Impl√©mentez et testez un algorithme r√©cursif qui r√©sout le probl√®me. Donnez une solution pour N=[1,10]

Pour tester votre solution ("est-ce que la solution retourne le nombre correct d'arbres?"), vous devez lire, ~comprendre, impl√©menter et tester la formule qui calcule le nombre de Catalan pour un N donn√©:

C<sub>n</sub> = $\frac{(2n)!}{(n+1)!n!}$

Aide: https://fr.wikipedia.org/wiki/Nombre_de_Catalan

```python
import math

# Fonction pour calculer le nombre de Catalan (Le nombre de Catalan compte le nombre d'arbres binaires complets (o√π chaque n≈ìud interne a exactement deux enfants) avec ùëõ  n≈ìuds non-feuilles)
def catalan_nb(n: int)->int:
    return math.comb(2*n,n)// (n+1)

catalan_nb(5)
```

```python
assert catalan_nb(3) == 5
assert catalan_nb(5) == 42
assert catalan_nb(8) == 1430
```

```python
# A COMPLETER
# Vous n'√™tes pas oblig√©s d'utiliser cette m√©thode avec cette signature. Vous pouvez cr√©er d'autres classes/m√©thodes
def get_all_B_Trees(n: int):
    if n == 0:
        return ['..']  # Cas de base: une feuille repr√©sente un arbre vide
    trees = []  # tableau pour stocker les arbres binaires g√©n√©r√©s
    for left_size in range(n): # Pour chaque taille possible de sous-arbre gauche
        right_size = n - 1 - left_size # Calculer la taille du sous-arbre droit
        left_trees = get_all_B_Trees(left_size) # Appel r√©cursif pour g√©n√©rer tous les arbres possibles pour la taille du sous-arbre gauche
        right_trees = get_all_B_Trees(right_size) # Appel r√©cursif pour g√©n√©rer tous les arbres possibles pour la taille du sous-arbre droit
        for left in left_trees: # Pour chaque arbre de gauche
            for right in right_trees: # Pour chaque arbre de droite
                trees.append(f'({left}{right})') # Combiner les arbres de gauche et de droite et les ajouter √† la liste finale des arbres binaires
    return trees

# Afficher tous les arbres binaires pour N=3 (pour tester)
print("Nombre de Catalan pour v√©rification:", catalan_nb(3))
get_all_B_Trees(3)
```

Qu'elle est la complexit√© en espace/temps de l'algorithme ? Une approximation asymptotique est recommand√©e. Vous aurez besoin de la formule de Stirling pour cela : $n ! \approx \sqrt{2 \pi n}\left(\frac{n}{\mathrm{e}}\right)^n$

### Explications

Time complexity:
We know that $C_n=\frac{1}{n+1}\binom{2 n}{n}$ ; $ \binom{2 n}{n} =\frac{(2 n)!}{n!}$ and applying Stirling's $n!\approx \sqrt{2 \pi n}\left(\frac{n}{e}\right)^n$.


$$
\begin{aligned}
& (2 n)!\approx \sqrt{4 \pi n}\left(\frac{2 n}{e}\right)^{2 n} \\
& (n!)^2 \approx(2 \pi n)\left(\frac{n}{e}\right)^{2 n} \\
& \binom{2 n}{n} \approx \frac{\sqrt{4 \pi n}\left(\frac{2 n}{e}\right)^{2 n}}{(2 \pi n)\left(\frac{n}{e}\right)^{2 n}} \approx \frac{2^{2 n}}{\sqrt{\pi n}}
\end{aligned}
$$


$$
\begin{aligned}
&c_n=\frac{1}{n+1} \cdot \frac{2^{2 n}}{\sqrt{\pi n}} \approx \frac{4^n}{n^{3 / 2} \sqrt{\pi}}\\
&\text { Time complexity is: } O\left(C_n\right) \approx O\left(\frac{4^n}{n^{3 / 2}}\right)
\end{aligned}
$$

$$
\begin{aligned}
&\text { Space complexity: }\\
&O\left(n \cdot c_n\right) \approx O\left(n \cdot \frac{4^n}{n^{3 / 2}}\right)=O\left(\frac{4^n}{\sqrt{n}}\right)
